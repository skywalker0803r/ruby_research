```
import matplotlib.pyplot as plt
import numpy as np

def f(x,a,c):
  return (abs(a-2)/a)*(((x/c)**2/abs(a-2)+1)**(a/2)-1)
```

```
x = np.linspace(-6,6,1000)
```

```
for a,l in zip([-1e+64,-2.0,1e-64,0.5,1,3/2,2+1e-8],['-âˆž','-2','0','1/2','1','3/2','2']):
  plt.plot(x,f(x,a=a,c=1),label='a={}'.format(l))
plt.legend(loc='upper right')
plt.ylim(0,5)
plt.show()
```

```
for a,l in zip([-1e+64,-2.0,1e-64,0.5,1,3/2,2+1e-8],['-âˆž','-2','0','1/2','1','3/2','2']):
  plt.plot(x,f(x,a=a,c=0.5),label='a={}'.format(l))
plt.legend(loc='upper right')
plt.ylim(0,5)
plt.show()
```
