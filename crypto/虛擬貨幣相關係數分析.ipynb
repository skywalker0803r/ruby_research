```
import finlab_crypto
finlab_crypto.setup()
```

```
import pandas as pd
table = pd.DataFrame()
for cname in ['BTC','BNB','ETH','XRP','LTC','DOGE']:
  table[cname] = finlab_crypto.crawler.get_all_binance(f'{cname}USDT','4h').close
table.tail()
```

```
((table-table.min())/(table.max()-table.min())).plot()
```

```
table.corr().style.background_gradient()
```

```
t = table[['BTC','DOGE']]
((t-t.min())/(t.max()-t.min())).plot()
```
